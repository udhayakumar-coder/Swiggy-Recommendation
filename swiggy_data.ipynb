```python
for i in df.columns:
    if df[i].isnull().sum():
        print(i, df[i].isnull().sum())
    else:
        print(i, "No null values")
```

```python
# Check for null values in particular columns
df = df.dropna(subset=['name', 'rating', 'rating_count', 'cost', 'cuisine'], how='all')
```

```python
for i in df.columns:
    if df[i].isnull().sum():
        print(i, df[i].isnull().sum())
    else:
        print(i, "No null values")
```

```python
df["cost"] = df["cost"].replace('₹', '', regex=True).str.strip().astype(int)
```

```python
df = df.dropna(subset=["cuisine"],how='any')
```

```python
for i in df.columns:
    if df[i].isnull().sum():
        print(i, df[i].isnull().sum())
    else:
        print(i, "No null values")
```

```python
# Replace '--' or other non-numeric values with 0, then convert
df["rating"] = df["rating"].replace('--', 0).astype(float)
```

```python
df[['Area', 'City']] = df['city'].apply(
   lambda x: x.rsplit(',', 1) if ',' in str(x) else ['', x]
).apply(pd.Series)
```

```python
df['Area'] = df['Area'].fillna('Unknown')       # NaN values நிரப்பும்
df['Area'] = df['Area'].replace('', 'Unknown')  # காலி string values நிரப்பும்
```

```python
import random
```

```python
# rating_count column has categorical values, we will convert them into numerical values
results = []
for i in df["rating_count"]:
    if i == "Too Few Ratings":
        results.append(random.randint(0, 20))
    elif i == '50+ ratings':
        results.append(random.randint(50, 100))
    elif i == '100+ ratings':
        results.append(random.randint(100, 200))
    elif i == '20+ ratings':
        results.append(random.randint(20, 50))
    elif i == '500+ ratings':
        results.append(random.randint(500, 1000))
    elif i == '1K+ ratings':
        results.append(random.randint(1000, 5000))
    elif i == '5K+ ratings':
        results.append(random.randint(5000, 10000))
    elif i == '10K+ ratings':
        results.append(random.randint(10000, 20000))
    else:
        results.append(i)
```

```python
df["rating_count"] = results
```

```python
from sklearn.preprocessing import LabelEncoder, MultiLabelBinarizer, OneHotEncoder
le=LabelEncoder()
mlb=MultiLabelBinarizer()
ohe=OneHotEncoder()
```

```python
# Apply LabelEncoder to name data
cl_df["name"]=le.fit_transform(df["name"])
```

```python
# Create and fit encoder
ohe = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
encoded = ohe.fit_transform(cl_df[['City', 'Area']])

# Convert encoded data to DataFrame (keep same index)
encoded_df = pd.DataFrame(
    encoded,
    columns=ohe.get_feature_names_out(['City', 'Area']),
    index=cl_df.index
)
```

```python
#Combine with original DataFrame
cl_df = pd.concat([cl_df, encoded_df], axis=1)
```

```python
cl_df.drop(columns=['City', 'Area'],axis=1, inplace=True)
```

```python
# 1. Split by comma (multiple sub-locations)
cl_df["cuisine"] = cl_df["cuisine"].apply(lambda x: [i.strip() for i in x.split(',')])

# 2. Apply MultiLabelBinarizer
mlb = MultiLabelBinarizer()
cuisine_new = mlb.fit_transform(cl_df["cuisine"])

# 3. Create new columns for each unique city/area
cuisine_encoded_df = pd.DataFrame(cuisine_new, columns=mlb.classes_, index=cl_df.index)

# 4. Combine with original DataFrame
cl_df = pd.concat([cl_df, cuisine_encoded_df], axis=1)
cl_df.drop(columns="cuisine",axis=1, inplace=True)
```
