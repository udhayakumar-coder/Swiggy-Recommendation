```python
# rating_count column has categorical values, we will convert them into numerical values
results = []
for i in df["rating_count"]:
    if i == "Too Few Ratings":
        results.append(random.randint(0, 20))
    elif i == '50+ ratings':
        results.append(random.randint(50, 100))
    elif i == '100+ ratings':   
        results.append(random.randint(100, 200))
    elif i == '20+ ratings':
        results.append(random.randint(20, 50))
    elif i == '500+ ratings':
        results.append(random.randint(500, 1000))
    elif i == '1K+ ratings':
        results.append(random.randint(1000, 5000))
    elif i == '5K+ ratings':
        results.append(random.randint(5000, 10000))
    elif i == '10K+ ratings':
        results.append(random.randint(10000, 20000))
    else:
        results.append(i)
```

```python
df["rating_count"] = results
```

```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder, MultiLabelBinarizer
ohe=OneHotEncoder()
le=LabelEncoder()
mlb=MultiLabelBinarizer()
```

```python
df["name"]=le.fit_transform(df["name"])
```

```python
df["name"].unique()
```

```python
# 1. Split by comma (multiple sub-locations)
df["city"] = df["city"].apply(lambda x: [i.strip() for i in x.split(',')])

# 2. Apply MultiLabelBinarizer
mlb = MultiLabelBinarizer()
city_new = mlb.fit_transform(df["city"])

# 3. Create new columns for each unique city/area
encoded_df = pd.DataFrame(city_new, columns=mlb.classes_)

# 4. Combine with original DataFrame
df = pd.concat([df, encoded_df], axis=1)

```
