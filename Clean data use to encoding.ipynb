```python
import pandas as pd
```

```python
df=pd.read_csv('/content/cleaned_data.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.set_index("id",inplace=True)
```

```python
df.head()
```

```python
df.shape
```

```python
from sklearn.preprocessing import LabelEncoder, MultiLabelBinarizer
le=LabelEncoder()
mlb=MultiLabelBinarizer()
```

```python
# Apply LabelEncoder to name data
df["name"]=le.fit_transform(df["name"])
```

```python
# 1. Split by comma (multiple sub-locations)
df["city"] = df["city"].apply(lambda x: [i.strip() for i in x.split(',')])

# 2. Apply MultiLabelBinarizer
mlb = MultiLabelBinarizer()
city_new = mlb.fit_transform(df["city"])

# 3. Create new columns for each unique city/area
encoded_df = pd.DataFrame(city_new, columns=mlb.classes_, index=df.index)

# 4. Combine with original DataFrame
df = pd.concat([df, encoded_df], axis=1)
df.drop(columns="city",inplace=True)
```

```python
# 1. Split by comma (multiple sub-locations)
df["cuisine"] = df["cuisine"].apply(lambda x: [i.strip() for i in x.split(',')])

# 2. Apply MultiLabelBinarizer
mlb = MultiLabelBinarizer()
cuisine_new = mlb.fit_transform(df["cuisine"])

# 3. Create new columns for each unique city/area
cuisine_encoded_df = pd.DataFrame(cuisine_new, columns=mlb.classes_, index=df.index)

# 4. Combine with original DataFrame
df = pd.concat([df, cuisine_encoded_df], axis=1)
df.drop(columns="cuisine",axis=1, inplace=True)
```
